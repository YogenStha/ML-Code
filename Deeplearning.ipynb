```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
m= 0.7
c= 0.3
x = np.arange(0,1,0.02)
y = m*x+c
print("shape of x= ",x.shape)
print("shape of y= ",y.shape)
```

```python
print (x[:5])
```

```python
split= int(0.8*len(x))
x_train,y_train= x[:40],y[:40]
x_test,y_test= x[40:],y[40:]
```

```python
def plot_data(x_train=x_train,y_train=y_train,x_test=x_test,y_test=y_test,predictions=None):
    plt.figure(figsize=(10,6))
    plt.scatter(x_train,y_train,c='b',s=8, label='Training Data')
    plt.scatter(x_test,y_test,c='r',s=8, label='Training Data')
    if predictions is not None:
        plt.scatter(x_test,predictions,c='g',s=8,label='Predicted data')
    plt.legend()
    plt.show()
```

```python
plot_data()
```

```python
y_preds = 0.2*x_test +0.55
```

```python
W = np.random.randn()/2
b = np.random.randn()/2
print("w = ",W)
print("b = ",b)
```

```python
def loss_function(y_true,y_preds):
    return np.sum(np.square(y_true-y_preds))/len(y_true)
```

```python
#Training
epochs = 500 #number of iterations
lr = 0.1 #learning rate
trainLossList = [] #to keep track of training loss
testLossList =[] # tp keep track of testing loss
for epoch in range(epochs):
    #Forward Pass
    y_preds = x_train*W +b
    #compute Train Loss
    train_loss = loss_function(y_true=y_train,y_preds=y_preds)
    trainLossList.append(train_loss)
    #compute the gradients
    gradientW =(-2*np.sum((y_train-y_preds)*x_train)/len(y_train))
    gradientb = (-2*np.sum(y_train-y_preds))/len(y_train)
    #Backprobagate i.e UPDATE THE PARAMETER
    W = W-lr*gradientW
    b = b-lr*gradientb
    #Compute the test loss
    test_preds = x_test*W+b
    test_loss = loss_function(y_true=y_test, y_preds=test_preds)
    testLossList.append(test_loss)
    #print what's happending..
    print(f'epoach: {epoch+1}| Train Loss: {train_loss:.4f}| Test Loss:{test_loss:.4f}')
```

```python
y_preds = x_test*W + b
plot_data(predictions=y_preds)
```

```python
#Plot Loss Curves
plt.figure(figsize =(10,6))
plt.plot(range(epochs),trainLossList,label="Training Loss")
plt.plot(range(epochs),testLossList,label="Testing Loss")
plt.title("Epochs vs Loss")
plt.xlabel("Epoche")
plt.ylabel("Loss")
plt.legend();
plt.savefig('loss.png')
plt.show()
```
