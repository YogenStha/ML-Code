```python
import pandas as pd
df = pd.read_csv("Life Expectancy Data.csv")
df.head()
null_ind = df[df["Adult Mortality"].isnull()].index
print(null_ind)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#create a new column in df to store encoded values
df["encoded_status"] = label_encoder.fit_transform(df["Status"])
print(f"{df[["Status","encoded_status"]][100:120]}")
```

```python
#imoprt Scaling from sklearn
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
#create object of MinMax Scalar
minmax_scalar =MinMaxScaler()
standard_scaler =StandardScaler()
```

```python
#Transforming the numerical value into a range of [0-1] ubing MinMax
#Make it easier for algorith to process these smaller values
df["scaled_Mortality"] = minmax_scalar.fit_transform(df[["Adult Mortality"]])
print(f"{df[["scaled_Mortality","Adult Mortality"]][10:30]}")
```

```python
#using standard scalar to transform numerical col
df["scaled_infant"] = standard_scaler.fit_transform(df[["infant deaths"]])
print(f"{df[["scaled_infant","infant deaths"]][10:30]}")
```

```python
#splitting the overall data into training and testing
#x = store feature columns
# y = store traget xolum
x= df.drop(columns =["Status","Country","Life expectancy "],axis=1)
y = df["Life expectancy "]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2, random_state= 42)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
y_predict = reg.predict(x_test)
print(y_predict)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae = mean_absolute_error(y_test,y_predict)
mse = mean_squared_error(y_test,y_predict)
r2 = r2_score(y_test,y_predict)
print(f"mean absolute error:{mae}")
print(f"mean square error:{mse}")
print(f"R-squared:{r2}")
```
