```python
#PCA= principle component analysis
from sklearn.datasets import load_iris
import pandas as pd
```

```python
data= load_iris()
data
```

```python
df =pd.DataFrame(data.data,columns=data.feature_names)
df['Species'] =data.target_names[data.target]
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df,hue='Species')
plt.show()
```

```python
#import PCA from sklearn.decomposition
#reduce the x to preserve top 3 princeple components
from sklearn.decomposition import PCA
x_reduced = PCA(n_components=3).fit_transform(data.data)
```

```python
#creating a 3d plot for principle components
fig  = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection = '3d',elev = -150, azim = 110)
scatter = ax.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2],c= data.target,s= 15)
#label and legends
ax.set(title ='PCA Dimension',xlabel= '1 Eigenvector',ylabel ='2 Eigenvector',zlabel ='3 Eigenvector')
legend1= ax.legend(scatter.legend_elements()[0],data.target_names,loc = 'upper right',title= 'Species')
ax.add_artist(legend1)
plt.show()
```

```python
#creating a 3d plot for principle components
fig  = plt.figure(figsize=(8,6))
ax = fig.add_subplot(projection = '3d',elev = 200, azim = 180)
scatter = ax.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2],c= data.target,s= 15)
#label and legends
ax.set(title ='PCA Dimension',xlabel= '1 Eigenvector',ylabel ='2 Eigenvector',zlabel ='3 Eigenvector')
legend1= ax.legend(scatter.legend_elements()[0],data.target_names,loc = 'upper right',title= 'Species')
ax.add_artist(legend1)
plt.show()
```
