```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
data= load_iris()
data
```

```python
df =pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
from sklearn.cluster import KMeans
#findind the number of clusters
wcss=[]#store score
x =df[['petal length (cm)','petal width (cm)']]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = 'k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.title('the  Elbow Method')
plt.show()
                    
                
```

```python
#as we now know the number of clusters = 3 from the elow graph we can use algo for
kmeans = KMeans(n_clusters=3, init='k-means++',random_state=0)
#predict the cluster for each data point 
kmeans.fit_predict(x)
```

#0,1,2 in above result refers to the cluster-0,cluster-1,cluster-2

```python
y_predict = kmeans.fit_predict(x)
y_predict
```

```python
#plotting the clusters
plt.scatter(x[y_predict==0]['petal length (cm)'],x[y_predict==0]['petal width (cm)'],c='r',label = "a")
plt.scatter(x[y_predict==1]['petal length (cm)'],x[y_predict==1]['petal width (cm)'],c='g',label = "b")
plt.scatter(x[y_predict==2]['petal length (cm)'],x[y_predict==2]['petal width (cm)'],c='b',label = "c")
plt.legend()
plt.show()
```

```python
#use sihouette_score to check the cluster score
from sklearn.metrics import silhouette_score
print(silhouette_score(x,y_predict))
```
